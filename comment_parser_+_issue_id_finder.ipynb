```
repos = []
f = open("repos.txt", "r")
for line in f:
  repos.append(line.replace("\n", ""))

print(repos)
```

```
import re

def getRepoName(url):
  pattern = "(?:git@|https://)github.com[:/](.*)"
  result = re.findall(pattern, url)
  if result:
    return result[0].split("/")[1]
  else:
    return None
```

```
import git

for url in repos:
  repo_name = getRepoName(url)
  git.Repo.clone_from(url,'repos/'+repo_name)
```

```
import ghlinguist as ghl
import glob
import os
import magic
from comment_parser import comment_parser
from comment_parser.parsers.common import UnterminatedCommentError
import traceback 
import sys
import csv
import time

folder = "maven"
start_time = time.time()
print(folder)
langs = ghl.linguist('repos/'+folder)
print("--- Analyze Language %s seconds ---" % (time.time() - start_time))
lang = langs[0][0]
#lang = 'Java'
ext = []
mime = ""
if lang == 'Java':
  ext.append(".java")
  mime = "text/x-java-source"
elif lang == 'Python':
  ext.append(".py")
  mime = 'text/x-python'
elif lang == 'JavaScript':
  ext.append(".js")
  ext.append(".jsx")
  mime = "application/javascript"

start_time = time.time()
files = []
for ftype in ext:
  files.extend(glob.glob('repos/'+folder+'/**/*'+ftype, recursive=True))
print("--- Capture Files %s seconds ---" % (time.time() - start_time))
print(len(files))

all_comments = {}
start_time = time.time()
for cfile in files:
  print(cfile)
  f = open(cfile, 'r')
  code = f.read()
  comments = comment_parser.extract_comments_from_str(code, mime="text/x-java-source")
  all_comments[cfile] = comments
print("--- Extract Comments %s seconds ---" % (time.time() - start_time))

print(len(all_comments))
```

```
import csv
with open('comments.csv', 'w', newline='') as file:
  writer = csv.writer(file)
  writer.writerow(["File", "Line", "Comment"])
  for file in all_comments.keys():
    for comment in all_comments[file]:
      writer.writerow([file, comment.line_number(), comment.text()])
```
